```python
#Create column for year, month, day, hour, minute, second, Type, Duration, Date
year=[]
month=[]
day=[]
hour=[]
minute=[]
second=[]
duration=[]
Type=[]
for i in range(len(data)):
    ligne=data.iloc[i]
    year.append(ligne['date'][0:4])
    month.append(ligne['date'][4:6])
    day.append(ligne['date'][6:8])
    hour.append(ligne['date'][9:11])
    minute.append(ligne['date'][11:13])
    second.append(ligne['date'][13:15])
    duration.append(float(ligne['Duration']))
    Type.append(ligne['Type'])
    
date=[]
for i in range(len(year)):
    date.append(datetime.datetime(year=int(year[i]), month=int(month[i]), day=int(day[i]), 
                                 hour=int(hour[i]), minute=int(minute[i]), second=int(second[i])))
    

d = {'year': year, 'month': month, 'day': day, 'hour': hour, 'minute': minute, 'second': second,
    'Type': Type, 'duration': duration, 'date': date}

cat=pd.DataFrame(data=d)
```

```python
#Save catalog
outputCat="ExampleCatalogue.pd"
pickle.dump(cat,open(outputCat,"wb"))
```
